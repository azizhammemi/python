##**<font color='red'>1-</font> $matplotlib$**
---

- Le module $matplotlib.pyplot$ est une partie de la bibliothèque $matplotlib$ qui est largement utilisée dans les domaines scientifiques et techniques pour la création de graphiques et la visualisation de données en Python.

- $matplotlib.pyplot$ fournit une interface pour créer des figures avec des axes, des zones de traçage, des étiquettes pour les axes, des titres et des légendes, et permet de produire une grande variété de types de graphiques tels que des courbes, des histogrammes, des diagrammes à secteurs, etc.










<u>**<font color='green'>Importation de bibliothèques**</u></font> <br>

```python
import numpy as np
import pandas as pd
#matplotlib
import matplotlib.pyplot as plt
#seaborn
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

---
### **<font color='red'> Visualisation avec pyplot : </font> <font color='blue'> plt.plot() / plt.show()</font>**
---



```python
plt.figure(figsize=(2,2))
aa1=range(5)
plt.plot(aa1) # Équivalent à plt.plot(aa1,aa1)
#plt.show()
 # Créer un plot en utilisant les valeurs d’axe x comprises entre 0 et 4 et les valeurs d’axe y aussi comprises entre 0 et 4
aa2= [9, 7, 8,6, 2]
plt.plot(aa1,aa2) # Créer un plot en utilisant les valeurs d’axe x comprises entre 0 et 4 et les valeurs d’axe y telles qu’elles apparaissent dans aa2
#plt.show() # Afficher la figure qui peut contenir de nombreux Axes (2 plt.plot(…)→ 2 Axes)


# NB. aa1 et aa2 doit être de la même longueur
```

---
### **<font color='red'> Définition de style et couleurs </font>**
---

```python
L1 = [1, 5, 8, 11, 2, 0, 7, 10, 4, 12]
L2 = [3, -8, -9, -2, -1, -2, -4, -7, -6, -12]

#couleur,style,marqueur
plt.plot(range(1,11), L1, 'r--o') # présenter une ligne pointillée (Dashed) en rouge dont les cordonnées des points de cette ligne sont marquées par un cercle
plt.plot(range(1,11), L2,'b:s') # présenter une ligne pointillée (Dotted) en bleu dont les cordonnées des points de cette ligne sont marquées par un carrée
plt.show()
```

---
### **<font color='red'> Définition titre pour les axes /   label aux axis (xlabel,ylabel)**
---


```python
l1 =range(8)
l2 = range(5,13)


#titre
plt.plot(l1,l2)
plt.title('Titre…', fontsize=20, fontname='Arial')

# Définir un label aux Axis
plt.xlabel('hahahhahaha')
plt.ylabel('eoooooooo')

plt.show()
```

---
### **<font color='red'> Définition des annotations et légendes**
---

```python
L12 = [1, 5, 8, 11 ,0, 7, 10 ,12]
L22 = [3, 8, 9, 2, 1, 4, 7, 6]

line1 = plt.plot(L12, L22)
line2 = plt.plot(L12, L12)


#annotations
plt.annotate('local max', xy=(8, 9), xytext=(7, 10), arrowprops=dict(facecolor='black'))  #arrowprops= flèche

plt.legend(['XXXX', 'YYYYY'], loc=0) #RQ. pour "loc": la position de la lègende peut être entre 0 et 10
plt.show()
```

---
### **<font color='red'> Définition des axes :</font> </font> <font color='blue'> set_xlim(...) / set_ylim(...)</font>**
---

```python
x1 = range(6)
x2 = [-15,  -27 ,  -3 ,  -1  ,  0  , 10]
plt.title('fig 1')
plt.plot(x1,x2)
plt.show()

#récupérer les axes
ax = plt.gca()

#Changer les limites (longueur) d’Axis x
ax.set_xlim([0, 11])
# Changer les limites d’Axis y
ax.set_ylim([-4, 5])

plt.title('fig 2')
plt.plot(x1,x2)
plt.show()
```

---
### **<font color='red'> Définition des ticks :</font> </font> <font color='blue'> set_xtick(...) / set_yticks(...)</font>**
---

```python
x = [1,2.7,3,4]
y = [1.4,2,3,4]

plt.figure(figsize=(3,2))
plt.title("Fig2")
plt.plot(x,y)
plt.show()

plt.figure(figsize=(3,2))
plt.plot(x,y)
ax=plt.gca()
ax.set_xticks([1,2.7,3,4],[1,2.7,3,4])

#plt.xticks([1,2.7,3,4],[1,2.7,3,4])
plt.show()
```

**EXEMPLE 1**

```python
x= np.arange(30)
y = 0.5 *x

fig, ax = plt.subplots(2)
ax[0].plot(x,y)
ax[1].plot(x,x**2)
ax[0].set_yticks([1,20,40])

#ax[0].set_title('XXXXXX')

plt.show()
```

**EXEMPLE 2**

```python
x = np.arange(0,40)
y = np.sin(x)
fig, ax = plt.subplots(2,3)
ax[0][0].plot(x, y)
ax[0][0].plot(y, x)
ax[0][1].plot(x)
ax[1][0].plot(y,x)
ax[0][0].set_title("ZZZZ")
plt.show()
```

**EXEMPLE 3**

```python
#*************************NB.**************************#
plt.figure(figsize=(10, 10)) #figsize = Width, height in inches. 1 in = 2.54 cm (default: [6.4, 4.8])
#*************************---**************************#

x = np.arange(0,40)
y = np.sin(x)
for i in range(4):
    ax=plt.subplot(2,2,i+1)
    plt.grid(True) #****
    plt.plot(x,y)
    plt.title("Courbe"+str(i+1))
plt.show()
```

**EXEMPLE 4** *plotting images*

```python
plt.figure(figsize=(5, 5))
for i in range(6):
    ax=plt.subplot(3,2,i+1)
    plt.imshow(np.eye(20),cmap=plt.cm.gray)
    plt.title("Image"+str(i+1))
    plt.axis("off")
plt.show()
```

###**$Seaborn$?**
- $seaborn$ est une bibliothèque Python pour la visualisation de données, basée sur la bibliothèque $matplotlib$ et intégrée aux structures de **Pandas**.
- $seaborn$  propose une variété de fonctions pour créer des graphiques tels que des histogrammes, des graphiques en barres, des nuages de points (scatter), etc.
- L'avantage de $seaborn$ par rapport à $matplotlib$ est sa capacité à simplifier la création de graphiques statistiques complexes grâce à ses fonctionnalités de haut niveau.

**Scatter avec $matplotlib$**

```python
x = df1['age']
y = df1['période de guérison (mois)']
plt.scatter(x,y, c='cyan', marker='^')
# par défaut c='#1f77b4' et marker='o'

plt.ylabel('période de guérison (mois)')
plt.xlabel('age')
plt.show()

#oubien
df1.plot.scatter(x='age',y='période de guérison (mois)')
plt.show()
```

**Scatter avec $Seaborn$**

```python
sns.scatterplot(data=df1, x="age", y="période de guérison (mois)")
plt.show()
```

---
### **<font color='red'> Histogrammes </font>**
---


```python
plt.figure(figsize=(10, 6))
h=plt.hist(df1["situation familiale"], bins=5, rwidth=0.5)
#color = 'red', edgecolor = 'blue'
positions = np.linspace(0.4, 3.6, 5) #?
print(positions)
labels=('divorcé', 'célibataire', 'marié', 'en couple', 'séparé')
plt.xticks(positions, labels)
plt.xlabel("situation familiale")
plt.ylabel("count")
plt.show()
```

**Histogramme avec $Seaborn$**

```python
plt.figure(figsize=(10, 6))

#sns.histplot(x="age", data=df1, bins=6)
sns.countplot(x="période de guérison (mois)", data=df1)
plt.show()
```
