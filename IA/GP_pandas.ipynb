**À partir d’un tableau ndarray**

### **<font color='red'> Opérations sur les dataframes </font>**
---

**Fusion des dataframes(<font color='blue'>pd.concat(...)</font>)**

```python
arr = np.array([[11,22,33],[44,55,66], [7,8,9], [99,88,33]])
df = pd.DataFrame(arr, columns = ['A','B','C'])

arr1 = np.array([[10,10,10],[12,12,12],[13,13,13]])

df1 = pd.DataFrame(arr1, columns = ['E','F','G'])

df2=pd.concat([df, df1], ignore_index=True)
df2
```

```python
df2=pd.concat([df, df1], axis=1) # ignore_index=True ?
df2
```

**Réinitialisation de l'index des lignes de dataframes(<font color='blue'>df.reset_index(...)</font>)**

```python
df1 = pd.DataFrame(arr1, columns = ['A','B','C'])

df2=pd.concat([df, df1])
df2.reset_index(drop=True, inplace=True)
df2

```

**Suppression des lignes/colonnes à partir d'un dataframe(<font color='blue'>df.drop(...)</font>)**

```python
#Supprimer les données de la colonne X2 de df

df.drop(labels=[2,3,4],axis=0, inplace=True)
df
```

```python
#Supprimer les lignes d'indice 1 et 3 de df2 (par défaut axis=0)
df=df.drop(index=[1,3])
```

**Tri sur les données d'un dataframe(<font color='blue'>df.sort_values(...)</font>)**

```python
arr = np.array([[11,22,33],[44,55,66], [7,8,9], [99,88,33]])
df = pd.DataFrame(arr, columns = ['A','B','C'])
df
```

```python
df.sort_values(by='B', inplace=True,ascending=False )
#df.sort_values(by='B', inplace=True,ascending=False, ignore_index=True)
df
```

```python
arr = np.array([[11,22,33],[44,55,66], [7,8,9], [99,88,33]])
df = pd.DataFrame(arr, columns = ['A','B','C'])
df.sort_values(by='B', inplace=True,ascending=False )
df.reset_index(drop=True, inplace=True)
df
```
