````python
biens=np.array([[85,120],[100,200],[85,150],[200,180], [100,420],[300,330],[400,400]])
````

**1/** G√©n√©rer un DataFrame √† partir de **biens**,$df\_biens$, poss√©dant $2$ colonnes, nomm√©es respectivement **surface** et **prix**. Afficher les $5$ premi√®res lignes de $df\_biens$

````python
df_biens = pd.DataFrame(biens, columns = ['surface','prix'])
df_biens
````

**2/** Afficher les surfaces des biens ayant un **prix** sup√©rieur ou √©gal √† $300$


````python
df_biens.loc[df_biens['prix'] >= 300, 'surface']
````

**3/** 	Cr√©er une nouvelle colonne, nomm√©e **total**, √† la fin de $df\_biens$ contenant les <u>prix totaux</u> des biens. Renommer la colonne **total** par **prixT**

````python
df_biens['total']=df_biens['surface']*df_biens['prix']
df_biens
````

**4/** 	Cr√©er une nouvelle colonne, nomm√©e **cat√©gorie**, √† la fin de $df\_biens$ qui prend la valeur **'B'** si **prixT** est inf√©rieur ou √©gal √† $32400$ et la surface est strictement inf√©rieure √† $120$, et **'A'** dans le cas contraire

````python
df_biens['categorie'] = np.where((df_biens['prix'] <= 32400) & (df_biens['surface']> 120), 'b','a')
df_biens
````

**1/** Charger le jeu de donn√©es **"athlete.csv"** dans un DataFrame nomm√© $df\_athlete$, puis afficher sa taille et les types de donn√©es de chaque attribut


````python
df_athlete=pd.read_csv('/content/drive/MyDrive/athlete.csv', sep=",")
print(df_athlete.shape)
df_athlete
````

**2/**	Tracer une courbe montrant le poids maximum des athl√®tes en fonction de leur **√¢ge**, <u>en marquant la valeur maximale des poids des athl√®tes ayant l‚Äô√¢ge de $16$ ans</u>

````python
#max=df_athlete.groupby(["Age"])["Weight"].max()
max=df_athlete[df_athlete["Age"] > 16].groupby("Age")["Weight"].max()

plt.figure(figsize=(5,3))
plt.plot(max.index,max.values)
plt.show()
````

**3/** Tracer un histogramme repr√©sentant la distribution des athl√®tes en fonction de $4$ intervalles d‚Äô**√¢ge**

````python
age_min = df_athlete["Age"].min()
age_max = df_athlete["Age"].max()
age_bins = np.linspace(age_min, age_max,5)
plt.xticks(age_bins)
print(age_bins)
sns.histplot(df_athlete["Age"], bins=age_bins)
plt.show()
````

**4/** Cr√©er une figure de dimensions $3\times11$ pouces (**in**) comprenant $2$ axes align√©s horizontalement, repr√©sentant <u>la distribution des athl√®tes selon leur sport</u>.
````python
plt.subplot(1,2,1)
sns.countplot(x = 'Sport',data=df_athlete)
plt.subplot(1,2,2)
df_athlete['Sport'].value_counts().plot.pie()
plt.tight_layout()

````
